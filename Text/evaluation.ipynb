### Sort by ascending order

```python
import pandas as pd
folder_path = '/content/drive/MyDrive/Jyothi Mam/Grading/averages_combined.csv'
df = pd.read_csv(folder_path)
df.head()
```

```python
# Remove '.csv' from the 'File Name' column
df['File Name'] = df['File Name'].str.replace('.csv', '')
# Remove 'P' from the 'File Name' column
df['File Name'] = df['File Name'].str.replace('P', '')

import re

# Function to extract numeric part from the 'File Name' column
def extract_numeric_part(s):
    match = re.search(r'\d+', s)
    if match:
        return int(match.group())
    return 0

# Sort the DataFrame by the numeric part of 'File Name' column
df_sorted = df.iloc[df['File Name'].astype(str).map(extract_numeric_part).sort_values().index]

# Display the sorted DataFrame
df_sorted.head()
```

```python
df_sorted.rename(columns={'File Name': 'Participant'}, inplace = True)
# print(df_sorted.head(10))
```

### Scaled to 10

```python
# List of columns to be scaled from 0-100 to 0-10
columns_to_scale = ['Fluency', 'Clarity', 'Confidence', 'Enthusiasm', 'Unique qualities', 'Professionalism', 'Engagement', 'Structured Answers', 'No Fillers', 'Focused', 'Authentic', 'Overall']

# Scale the specified columns to a 0-10 scale
df_sorted[columns_to_scale] = df_sorted[columns_to_scale] / 10

# Round all columns in the DataFrame to a single digit
rounded_value = df_sorted.round(0).astype(int)



# Display the updated DataFrame
rounded_value
```

```python
import os

# Define the path to the new folder
new_folder_path = '/content/drive/MyDrive/Jyothi Mam/Grading/Scaled'

# Check if the folder already exists, if not, create it
if not os.path.exists(new_folder_path):
    os.makedirs(new_folder_path)
    print(f"New folder '{new_folder_path}' created.")
else:
    print(f"Folder '{new_folder_path}' already exists.")

# Define the path for the output CSV file within the new folder
output_csv_path = os.path.join(new_folder_path, 'scaled_output.csv')

# Save the updated DataFrame to a CSV file within the new folder
rounded_value.to_csv(output_csv_path, index=False)
print(f"DataFrame saved to '{output_csv_path}'.")
```
